```python
import pickle

with open('./game_state.pkl', 'rb') as f:
    # Loading the quantum states generated at each step of the game process
    game_state = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
import imageio


import matplotlib.pyplot as plt


for key, item in game_state['game_step'].items():
    print(key)
    save_path = f'../images/game_step_{key}.png'
    # Draw and save images
    item.draw_game(save_path=save_path)
    
# Synthesize GIF files
images = []
for key in game_state['game_step'].keys():
    image_path = f'../images/game_step_{key}.png'
    images.append(imageio.imread(image_path))

gif_path = '../images/game_steps.gif'

imageio.mimsave(gif_path, images, duration=6.0)
```

```python
from IPython.display import display, Image, clear_output
import time

gif_path = '../images/game_steps.gif'

# Play GIF in a loop 5 times
for i in range(5):
    display(Image(filename=gif_path))
    time.sleep(4)
    # Clear output for next display
    clear_output(wait=True) 
```
